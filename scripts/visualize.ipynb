```python
import sys
sys.path.append('../')
from datasets.nordland import Nordland
import setup
import matplotlib.pyplot as plt
from torchvision.utils import make_grid
from PIL import Image
import numpy as np
import torch
```

```python
from setup import config

dataset = Nordland(config, config.train_data_path)

data_loader = torch.utils.data.DataLoader(dataset, batch_size=1, shuffle=True)

for anc, pos, neg in data_loader:
    break

combined_grid = torch.cat([anc, pos, neg], dim=0)
grid = make_grid(combined_grid, nrow=3)  

np_grid = grid.numpy().transpose((1, 2, 0))  
np_grid = np_grid - np.min(np_grid)
np_grid = np_grid / np.max(np_grid) 


plt.figure(figsize=(10, 5))
plt.imshow(np_grid)
plt.title('Visualisation of Anc, Pos, Neg Images')
plt.axis('off')
plt.show()
```
